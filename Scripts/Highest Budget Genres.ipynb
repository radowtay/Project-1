# Importing and Cleaning

```python
#import libraries

import pandas as pd
import matplotlib.pyplot as plt
```

```python
#reference new csv
file="../clean_data.csv"
df=pd.read_csv(file)

df=df.drop(columns="Unnamed: 0")
```

```python
df.head()
```

```python
#Pull out columns; edit type to fix budget errors
hbg_df=df[["Title", "Genre", "Budget"]]
hbg_df["Budget"].astype(int)
hbg_df.head(20)
```

```python
#prepare to search for and drop duplicates
dupsearch=hbg_df
```

```python
dupsearch_df=dupsearch.drop_duplicates(subset="Title", keep="first")
```

```python
dupsearch_df["Budget"] = pd.to_numeric(dupsearch_df["Budget"])
dupsearch_df
```

```python

dupsearch_df.sort_values(by='Budget')
```

# Split multi-genres into separate genres

```python
genre_df= dupsearch_df
genre_df['Genre']=genre_df['Genre'].str.split('|')
genre_df = dupsearch_df.explode('Genre')
genre_df.head()
```

```python
#Group by genre
genre_data_df = genre_df.groupby('Genre')

genre_data_df.count().sort_values(by="Title",ascending=False)
```

```python
budget_data_df=genre_df.groupby("Genre").mean()
budget_data_df['Budget'] = budget_data_df['Budget'].astype(int)
```

```python
budget_data_df
```

# Visualizations

## Average Budgets by Genre

```python
genres=budget_data_df.index
budget=budget_data_df.values
plt.title("Average Budgets by Genre", y=1.2)
plt.pie(budget, labels=genres, autopct='%1.1f%%', shadow=True, radius=1.3)
plt.savefig("../Images/Average_Budgets_by_Genre")
plt.show()
```

```python
budget_data_df.sort_values(by="Budget", ascending=False)
```

```python
top_budgets_df=budget_data_df[(budget_data_df.index =='Animation')|(budget_data_df.index =='Adventure')|
                   (budget_data_df.index =='Sci-Fi')|(budget_data_df.index=='Action')|
                   (budget_data_df.index=='Fantasy')|(budget_data_df.index=='Family')|
                   (budget_data_df.index=='War')]
```

## Highest Budgets by Genre

```python
genres=top_budgets_df.index
budget=top_budgets_df.values
explode=(0,0,0.2,0,0,0,0)
plt.title("Budget Comparisons for Highest Budget Genres", y=1)
plt.pie(budget, labels=genres, autopct='%1.1f%%', explode=explode, shadow=True, radius=1)
plt.savefig("../Images/IceCreamSales.png")
plt.show()
```

```python
top_budgets_df
```

```python
bottom_budgets_df=budget_data_df[(budget_data_df.index =='Romance')|(budget_data_df.index =='Sport')|
                   (budget_data_df.index =='Biography')|(budget_data_df.index=='Music')|
                   (budget_data_df.index=='Horror')|(budget_data_df.index=='Documentary')|
                   (budget_data_df.index=='Film-Noir')]
```

## Lowest Budgets by Genre

```python
genres=bottom_budgets_df.index
budget=bottom_budgets_df.values
explode=(0,0,0,0,0,0.2,0)
plt.title("Budget Comparisons for Lowest Budget Genres", y=1)
plt.pie(budget, labels=genres, autopct='%1.1f%%', explode=explode, shadow=True, radius=1)
plt.savefig("../Images/Lowest_Genre_Budgets.png")
plt.show()
```
